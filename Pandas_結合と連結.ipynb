## 連結と結合
### 連結
* ２つのDataFrameをさまざまな方法で結合する。
* inner join、outer joinっぽいこともできるが、行名をキーにするので、行名を意識して設定しているデータ（例えば、日別でのデータで、行名が日付）とかでないと使わないかも
* 同じレコード数の２つのDataFrameを結合する、もしくは行追加するときなどに使用するイメージかと思う。
* 縦方向の連結だけならappendの方がシンプル。ループ内でappendは使わないほうがよい？（引用先に記載されている）

```python
import os
import pandas as pd

df1 = pd.read_csv("test.csv",index_col='ind')
df1
```

```python
df2 = pd.read_csv("test2.csv",index_col='ind')
df2
```

```python
df3 = pd.concat([df1,df2])
df3
```

```python
df4 = pd.concat([df1,df2],axis=1)
df4
```

```python
df5 = pd.concat([df1,df2],axis=1,join='inner')
df5
```

```python
df6 = pd.concat([df1,df2],axis=1,join='outer')
df6
```

## 結合
* howで結合方法を指定可能。デフォルトはinner


```python
pf7 = pd.merge(df1,df2)
df7
```

```python
pf8 = pd.merge(df1,df2,how='left')
pf8
```

```python
df9 = pd.merge(df1,df2,how='right')
df9
```

```python
df10 = pd.merge(df1,df2,how='outer')
df10
```
